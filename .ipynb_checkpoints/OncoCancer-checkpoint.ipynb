# IMPORTING LIBRARIES

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import RFE
from sklearn.svm import SVC
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.metrics import accuracy_score
import xgboost as xgb  
from sklearn.linear_model import RidgeClassifier, Perceptron
from sklearn.linear_model import PassiveAggressiveClassifier
import pandas as pd
from scipy import stats
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score, classification_report
import matplotlib.pyplot as plt
import seaborn as sns

```

# LOADING DATASET

```python
dataset_folder = 'Datasets'
filename1 = f'{dataset_folder}/data11tumors2.csv'
data = pd.read_csv(filename1)
```

# Performing T-Test

```python
# Split data for class 1 and class 0
class_1_data = data[data['Class'] == 1].drop(columns=['Class'])
class_0_data = data[data['Class'] == 0].drop(columns=['Class'])

# Perform independent t-test for each gene
p_values = []
for gene in class_1_data.columns:
    t_stat, p_value = stats.ttest_ind(class_1_data[gene], class_0_data[gene], equal_var=False)
    p_values.append(p_value)

# Store the results in a DataFrame
t_test_results = pd.DataFrame({'Gene': class_1_data.columns, 'P-value': p_values})

# Sort by smallest p-value
t_test_results_sorted = t_test_results.sort_values(by='P-value')

# Show the top 10 differentially expressed genes
print("Top 10 differentially expressed genes between Class 1 and Class 0:")
print(t_test_results_sorted.head(10))
```

# DATA DESCRIPTION

```python
# Describe the dataset
print("Dataset Overview:")
print(f"Number of rows (samples): {data.shape[0]}")
print(f"Number of columns (genes + class): {data.shape[1]}")
print("\nFirst 5 rows of the dataset:")
print(data.head())

# Check for missing values
print("\nCheck for missing values:")
print(data.isnull().sum())

# Display basic statistics for numerical columns
print("\nStatistical summary of gene expression values:")
print(data.describe())

# Count the number of samples in each class
print("\nClass distribution (0: Normal, 1: Tumor):")
print(data['Class'].value_counts())
```

# FEATURE EXTRACTION

```python
# Encode the labels
labels = data['Class']
gene_data = data.drop(columns=['Class'])

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(gene_data, labels, test_size=0.3, random_state=42)

# Create a Support Vector Classifier (SVC)
svc = SVC(kernel="linear")

# Apply Recursive Feature Elimination (RFE)
rfe = RFE(estimator=svc, n_features_to_select=100, step=10)  # Adjust n_features_to_select as needed
rfe.fit(X_train, y_train)

# Get the ranking of features
ranking = rfe.ranking_

# Store RFE results
rfe_results = pd.DataFrame({'Gene': gene_data.columns, 'Ranking': ranking})

# Show the top-ranked features (smallest ranks are most important)
top_genes = rfe_results[rfe_results['Ranking'] == 1]
print("Top genes selected by RFE:")
print(top_genes)
```

# METRICS

```python
# Define a function to collect the metrics and confusion matrices
def collect_metrics(y_true, y_pred, classifier_name):
    acc = accuracy_score(y_true, y_pred)
    f1 = f1_score(y_true, y_pred, average='weighted')
    cm = confusion_matrix(y_true, y_pred)
    
    accuracy_scores.append(acc)
    f1_scores.append(f1)
    confusion_matrices.append(cm)
    
    # Also print the report for each classifier
    print(f"\n{classifier_name} Evaluation:")
    print(f"Accuracy: {acc:.4f}")
    print(f"F1 Score (Weighted): {f1:.4f}")
    print(f"Confusion Matrix:\n{cm}")
    print(f"Classification Report:\n{classification_report(y_true, y_pred, zero_division=0)}")

# Initialize lists for storing metrics
classifiers = ["SVM", "Logistic Regression", "LDA", "Perceptron", "Decision Tree", 
               "GaussianNB", "Random Forest", "Ridge", "Passive Aggressive", "XGBoost"]

accuracy_scores = []
f1_scores = []
confusion_matrices = []
```

# 1. SUPPORT VECTOR MACHINE

```python
# 1. Support Vector Machine Classifier
svm_model = SVC(kernel='linear', random_state=42)
svm_model.fit(X_train, y_train)
y_pred_svm = svm_model.predict(X_test)
collect_metrics(y_test, y_pred_svm, "SVM")
```

# 2. LOGISTIC REGRESSION

```python
# 2. Logistic Regression
log_reg_model = LogisticRegression(max_iter=1000, random_state=42)
log_reg_model.fit(X_train, y_train)
y_pred_log_reg = log_reg_model.predict(X_test)
collect_metrics(y_test, y_pred_log_reg, "Logistic Regression")
```

# 3. LINEAR Discriminant Analysis (LDA)

```python
# 3. Linear Discriminant Analysis
lda_model = LinearDiscriminantAnalysis()
lda_model.fit(X_train, y_train)
y_pred_lda = lda_model.predict(X_test)
collect_metrics(y_test, y_pred_lda, "LDA")
```

# 4. PERCEPTRON

```python
# 4. Perceptron
perceptron_model = Perceptron(random_state=42)
perceptron_model.fit(X_train, y_train)
y_pred_perceptron = perceptron_model.predict(X_test)
collect_metrics(y_test, y_pred_perceptron, "Perceptron")
```

# 5. Decision Tree Classifier

```python
# 5. Decision Tree Classifier
decision_tree_model = DecisionTreeClassifier(random_state=42)
decision_tree_model.fit(X_train, y_train)
y_pred_dt = decision_tree_model.predict(X_test)
collect_metrics(y_test, y_pred_dt, "Decision Tree")
```

# 6. Gaussian Naive Bayes

```python
# 6. Gaussian Naive Bayes
gnb_model = GaussianNB()
gnb_model.fit(X_train, y_train)
y_pred_gnb = gnb_model.predict(X_test)
collect_metrics(y_test, y_pred_gnb, "GaussianNB")
```

# 7. Random Forest Classifier

```python
# 7. Random Forest Classifier
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)
y_pred_rf = rf_model.predict(X_test)
collect_metrics(y_test, y_pred_rf, "Random Forest")
```

# 8. Ridge Classifier

```python
# 8. Ridge Classifier
ridge_model = RidgeClassifier(random_state=42)
ridge_model.fit(X_train, y_train)
y_pred_ridge = ridge_model.predict(X_test)
collect_metrics(y_test, y_pred_ridge, "Ridge")
```

# 9. PassiveAggressiveClassifier

```python
# 9. PassiveAggressiveClassifier
pac_model = PassiveAggressiveClassifier(max_iter=1000, random_state=42)
pac_model.fit(X_train, y_train)
y_pred_pac = pac_model.predict(X_test)
collect_metrics(y_test, y_pred_pac, "Passive Aggressive")
```

# 10. XGBoost Classifier

```python
# 10. XGBoost Classifier
xgb_model = xgb.XGBClassifier(random_state=42)
xgb_model.fit(X_train, y_train)
y_pred_xgb = xgb_model.predict(X_test)
collect_metrics(y_test, y_pred_xgb, "XGBoost")
```

```python
# Plot confusion matrices
def plot_confusion_matrices(classifiers, confusion_matrices):
    fig, axes = plt.subplots(2, 5, figsize=(20, 10))
    axes = axes.ravel()

    for i, (classifier_name, cm) in enumerate(zip(classifiers, confusion_matrices)):
        sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[i])
        axes[i].set_title(f'{classifier_name} Confusion Matrix')
        axes[i].set_xlabel('Predicted')
        axes[i].set_ylabel('True')

    plt.tight_layout()
    plt.show()

# Call the function to plot confusion matrices
plot_confusion_matrices(classifiers, confusion_matrices)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

def plot_metrics(classifiers, accuracy_scores, f1_scores):
    # Ensure that all lists have the same length
    assert len(classifiers) == len(accuracy_scores) == len(f1_scores), "Array lengths are not the same"

    fig, ax = plt.subplots(1, 2, figsize=(14, 5))

    # Accuracy plot
    sns.barplot(x=accuracy_scores, y=classifiers, hue=classifiers, ax=ax[0], palette="Blues_d", legend=False)
    ax[0].set_title("Accuracy of Classifiers")
    ax[0].set_xlabel("Accuracy")

    # F1 Score plot
    sns.barplot(x=f1_scores, y=classifiers, hue=classifiers, ax=ax[1], palette="Greens_d", legend=False)
    ax[1].set_title("F1 Score of Classifiers")
    ax[1].set_xlabel("F1 Score")

    plt.tight_layout()
    plt.show()

# Call the function to plot the metrics
plot_metrics(classifiers, accuracy_scores, f1_scores)
```
